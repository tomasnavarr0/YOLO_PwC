```python
import ultralytics
from ultralytics import YOLO


model = YOLO('yolov8n.pt') 

model.train(data='/kaggle/input/dataset-sr/dataset.yaml',  
            epochs=100,  
            imgsz=640,  
            batch=32,  
            name='yolov8_sr') 


results = model.val()


model.save('best_model.pt')
```
